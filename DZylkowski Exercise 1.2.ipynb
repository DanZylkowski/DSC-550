#### 1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
# Import numpy and pandas 
import numpy as np
import pandas as pd

# We set the random see for reproducible results
np.random.seed(0)

# Use the np.random() function to create a (3,4) test array of random integers between 1 and 10 inclusive
rand_test_array = np.random.randint(1,11, size=(3,4))

# Next we create a (10,50) array of random integers between 1 and 500 inclusive
rand_array = np.random.randint(1,501, size=(10,50))

# Load the arrays into dataframes
df_test = pd.DataFrame(rand_test_array)
df = pd.DataFrame(rand_array)
```

#### 2. Calculate the sum of each row; the sum of each column; and the sum of all entries of the test df.

```python
# We will first test our code on the test df
df_test
```

```python
# We will sum each row in the test DataFrame by summing the df along axis=1:
test_row_sum = df_test.sum(axis=1)

# We will sum each column in the test DataFrame by summing the df along axis=0:
test_column_sum = df_test.sum(axis=0)
```

```python
for i,j in enumerate(test_row_sum):
    print('The sum of row',i+1, '=',j)
```

```python
for i,j in enumerate(test_column_sum):
    print('The sum of column',i+1, '=',j)
```

```python
# Next we will sum the all values in the test df
print('The sum of all entries =',df_test.values.sum())
```

The code works on the test df so we will repeat the above using the actual df.  

#### 2. Calculate the sum of each row; the sum of each column; and the sum of all entries of the actual df. 

```python
# We can sum each row in the df by summing the df along axis=1:
row_sum = df.sum(axis=1)

# We can sum each column in the DataFrame by summing the df along axis=0:
column_sum = df.sum(axis=0)
```

```python
for i,j in enumerate(row_sum):
    print('The sum of row',i+1, '=',j)
```

```python
for i,j in enumerate(column_sum):
    print('The sum of column',i+1, '=',j)
```

```python
# Next we will sum the all values in the test df
print('The sum of all entries =',df.values.sum())
```

#### 3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
# We can use the numpy methods with the original random array 
x = np.max(rand_array)
y =np.min(rand_array)
z = np.mean(rand_array)
b = np.std(rand_array)

print ("Max =", x, " Min =", y, " Average =", z, " Standard deviation =", b)
```

```python
# We could also dataframe methods, as shown in the following comparison statements
x==df.values.max(), y==df.values.min(), z==df.values.mean(), b==df.values.std()
```
